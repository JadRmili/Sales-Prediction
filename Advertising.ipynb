# Librairies:

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
from sklearn.metrics import r2_score
```

# Import csv :

```python
df = pd.read_csv(r"C:\Users\user\Desktop\projet python\advertising.csv")
df = df.dropna()
```

## Tv Correlation:

```python
X_tv=df['TV'].values
Y_Sales=df['Sales'].values

#plot:
plt.figure(figsize=(20,10))
plt.scatter(X_tv,Y_Sales,label='Real')

#Regression:
reg =linear_model.LinearRegression()
reg.fit(df[['TV']],Y_Sales)

#prediction from Tv _column:
df['Prediction_Tv_Sales']=reg.predict(df[['TV']])

#Tracer la courbe de regression lineaire :
plt.plot(X_tv,df['Prediction_Tv_Sales'],color='red',label='predicted')
plt.legend()
plt.xlabel('Tv ADS')
plt.ylabel('Sales')
print('R squared = ',r2_score(Y_Sales,df['Prediction_Tv_Sales']))

```

## Radio Correlation:


```python
X_radio=df['Radio'].values
Y_Sales=df['Sales'].values

#plot:
plt.figure(figsize=(20,10))
plt.scatter(X_radio,Y_Sales)

#Regression:
reg =linear_model.LinearRegression()
reg.fit(df[['Radio']],Y_Sales)

#prediction from Tv _column:
df['Prediction_Radio_Sales']=reg.predict(df[['Radio']])

#Tracer la courbe de regression lineaire :
plt.plot(X_radio,df['Prediction_Radio_Sales'],color='red')
plt.xlabel('Radio ADS')
plt.ylabel('Sales')
print('R squared = ',r2_score(Y_Sales,df['Prediction_Radio_Sales']))
```
